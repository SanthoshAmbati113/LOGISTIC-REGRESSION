```python
import zipfile

with zipfile.ZipFile('/content/breast-cancer-wisconsin-data.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')  # or 'content/' based on your platform
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/data.csv')
df = df.drop(columns=['Unnamed: 32'])
```

```python
df.describe()
```

```python
df.sample(5)
```

```python
X=df.drop('diagnosis',axis=1)
```

```python
Y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
```

```python
X_train=scaler.transform(x_train)
X_test=scaler.transform(x_test)
```

```python
X_train = pd.DataFrame(X_train, columns=x_train.columns)
X_test = pd.DataFrame(X_test, columns=x_test.columns)
```

```python
X_train.shape
```

```python
df.shape
```

```python
df.loc[x_train.index,x_train.columns] = X_train
df.loc[x_test.index,x_test.columns] = X_test
```

```python
df.head()
```

```python
import numpy as np
np.round(df.describe())
y_train_num = y_train.map({'B': 0, 'M': 1})
y_test_num = y_test.map({'B': 0, 'M': 1})
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
Y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
print(accuracy_score(y_test,Y_pred))
print(confusion_matrix(y_test,Y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class
y_scores = lr.predict_proba(X_test)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test_num, y_scores)

# Calculate AUC
auc_score = roc_auc_score(y_test_num, y_scores)
print(f"AUC Score: {auc_score:.4f}")

# Plot ROC
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.4f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='grey')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
